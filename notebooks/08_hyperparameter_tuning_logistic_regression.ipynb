```python
import numpy as np
# Loading preprocessed data
X_train_scaled = np.load("data/X_train_scaled.npy")
X_test_scaled = np.load("data/X_test_scaled.npy")
y_train = np.load("data/y_train.npy")
y_test = np.load("data/y_test.npy")
X_train = np.load("data/X_train_unscaled.npy")
X_test = np.load("data/X_test_unscaled.npy")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold

lr = LogisticRegression(
    class_weight="balanced",
    max_iter=1000,
    n_jobs=-1
)

param_dist_lr = {
    "C": np.logspace(-3, 2, 10)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

rs_lr = RandomizedSearchCV(
    estimator=lr,
    param_distributions=param_dist_lr,
    n_iter=10,
    scoring="roc_auc",
    cv=cv,
    random_state=42,
    n_jobs=-1
)

rs_lr.fit(X_train_scaled, y_train)

print("Best LR parameters:", rs_lr.best_params_)
print("Best LR CV ROC-AUC:", rs_lr.best_score_)
```
